Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """

    return cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

```python
import os
import statistics as stats
imglist = os.listdir("test_images/")
def find_lines(image):

    #Greyscale
    gray = grayscale(image)
    
    #Blur
    kernel_size = 5
    blur_gray = gaussian_blur(gray, kernel_size)
    
    #Canny edge transformation
    low_threshold = 0
    high_threshold = 200
    edges = canny(blur_gray, low_threshold, high_threshold)
    
    #Create masked edges
    mask = np.zeros_like(edges)
    ignore_mask_color = 255
    imshape = image.shape
    vertices = np.array([[(110,imshape[0]),(410, 310),(480, 310), (imshape[1],imshape[0])]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)
    
    #Hough transform
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 50     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 100 #minimum number of pixels making up a line
    max_line_gap = 100    # maximum gap in pixels between connectable line segments
    
    lines = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)
    left_lines, right_lines = [],[]
    
    #Split lines into left and right sides
    for line in lines:
        x1,y1,x2,y2 = line[0]
        slope = (y2-y1)/(x2-x1)
        if slope < 0:
            left_lines.append(line)
        else:
            right_lines.append(line)
            
    for line_set in (left_lines, right_lines):
        if not line_set:
            continue
        
        slopes, y_intercepts = [], []
        for line in line_set:
            x1,y1,x2,y2 = line[0]
            slope = (y2-y1)/(x2-x1)
            y_intercept = y2-slope*x2
            if slope == float('inf'):
                continue
            slopes.append(slope)
            y_intercepts.append(y_intercept)
        
        #Calculate std dev and slope for line sets
        slope_mean = stats.mean(slopes)
        if len(slopes) == 1:
            slope_stdev = 0
        else:
            slope_stdev = stats.stdev(slopes)
        y_int_mean = stats.mean(y_intercepts)
        if len(y_intercepts) == 1:
            y_int_stdev = 0
        else:
            y_int_stdev = stats.stdev(y_intercepts)
        slope_low, slope_high = slope_mean - slope_stdev, slope_mean + slope_stdev
        y_int_low, y_int_high = y_int_mean - y_int_stdev, y_int_mean + y_int_stdev
        
        filtered_line_set = []
        for line in line_set:
            x1,y1,x2,y2 = line[0]
            slope = (y2-y1)/(x2-x1)
#             print('slope:', slope, 'mean:', slope_mean)
            y_intercept = y2-slope*x2
            
            #Check if slope and y intercept is within 1 stdev
            if (y_int_low <= y_intercept <= y_int_high) and (slope_low <= slope <= slope_high):
                filtered_line_set.append(line)
                
        #Figure out which line to show
#         print('original:', len(line_set))
#         print('filtered:', len(filtered_line_set))

        average_line = [[stats.mean(points[0]) for points in np.transpose(filtered_line_set)]]
#         print('avg', average_line)
        x1,y1,x2,y2 = average_line[0]
        slope = (y2 - y1)/(x2 - x1)
        if slope == 0:
            slope += .000001
        
        #Using point-slope calculations, when y=0, x= x1 - (y1/m)
#         print('info', x1, y1, x2, y2, slope)
        x_intercept = int(x1 + ((imshape[0]-y1)/slope))
        used_line = [[x_intercept, imshape[0], x2, y2] if slope < 0 else [x1, y1, x_intercept, imshape[0]]]
        
        line_set.clear()
        line_set.append(used_line)
                
        
    line_img = np.zeros((masked_edges.shape[0], masked_edges.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, left_lines + right_lines, thickness=5)
    return line_img
    
```

I decided to filter out noise using the mean and standard deviation. I made it so that any lines detected that were outside one standard deviation in either mean or y intercept were excluded. This worked fantastically. I then calculated the slope of each line to find where it hit the end of the picture so that the line length was a little more consistent. 

To improve, I could give it a constant line length, because right now the end of the line jumps around. The biggest problem I see with that right now is that the length would be pretty arbitrary, as there's no obvious place for the line to end. I might add that in later if it becomes necessary, but for the purposes of this project, I'm not sure that would add any information, it would simply make the current picture look a little prettier.


My current algorithm occasionally does not produce a line, but that's rare enough that the car could compensate by simply driving the direction if was previously going until it picks up a line again.

The pipeline takes a picture, makes it greyscale, finds lines, then isolates those lines into a triangle that's half the height of the picture with the corners being center and the two bottom corners. It then goes through the remaining lines, finds the mean and standard deviation of each the slopes and the y-intercepts, and erases all that are not within one standard deviation of the mean. It then takes the remaining lines, finds the average of each of the points, then uses those points to make a new average line, which is then extended toward the bottom of the picture.
